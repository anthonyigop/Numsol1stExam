```python
#PRoblem 1

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**5 + 3*x**3 - 10

def df(x):
    return 5*x**4 + 9*x**2

def bisection(a, b, tol=1e-6):
    if f(a) * f(b) > 0:
        return None
    while abs(b - a) > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

def newton(x0, tol=1e-6):
    while abs(f(x0)) > tol:
        x0 = x0 - f(x0) / df(x0)
    return x0

def secant(x0, x1, tol=1e-6):
    while abs(f(x1)) > tol:
        x_temp = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        x0, x1 = x1, x_temp
    return x1

root_bisection = bisection(1, 2)
root_newton = newton(1.5)
root_secant = secant(1, 2)

print("Bisection Root:", root_bisection)
print("Newton Root:", root_newton)
print("Secant Root:", root_secant)
```

```python
#Problem 2

import math

def f(x):
    return 8 * math.exp(7.3*x) + 3.892*x**3 - 10.3

def df(x):
    return 8 * 7.3 * math.exp(7.3*x) + 3 * 3.892 * x**2

def bisection(a, b, tol=1e-6):
    if f(a) * f(b) > 0:
        return None
    while abs(b - a) > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

def newton(x0, tol=1e-6):
    while abs(f(x0)) > tol:
        x0 = x0 - f(x0) / df(x0)
    return x0

def secant(x0, x1, tol=1e-6):
    while abs(f(x1)) > tol:
        x_temp = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        x0, x1 = x1, x_temp
    return x1

root_bisection = bisection(0, 1)
root_newton = newton(0.5)
root_secant = secant(0, 1)

print("Bisection Root:", root_bisection)
print("Newton Root:", root_newton)
print("Secant Root:", root_secant)

```

```python
#Problem 3

import numpy as np

def f(x):
    try:
        return np.real(x**7.23 * np.cos(x**3) + 25.8*x)
    except:
        return np.nan

def df(x):
    try:
        return np.real(7.23 * x**6.23 * np.cos(x**3) - 3 * x**9.23 * np.sin(x**3) + 25.8)
    except:
        return np.nan

def bisection(a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        return None

    for _ in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c

def newton_raphson(x0, tol=1e-6, max_iter=100):
    x = x0
    for _ in range(max_iter):
        if abs(f(x)) < tol:
            return x
        if df(x) == 0:
            return None
        x_new = x - f(x) / df(x)
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    return x

def secant(x0, x1, tol=1e-6, max_iter=100):
    for _ in range(max_iter):
        if abs(f(x1)) < tol:
            return x1
        if f(x1) - f(x0) == 0:
            return None
        x_new = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        if abs(x_new - x1) < tol:
            return x_new
        x0, x1 = x1, x_new
    return x1

a, b = -2, 2
x0 = 1.5
x1 = 2.0

root_bisection = bisection(a, b)
root_newton = newton_raphson(x0)
root_secant = secant(x0, x1)

print("Bisection Method Root:", root_bisection)
print("Newton-Raphson Method Root:", root_newton)
print("Secant Method Root:", root_secant)
```
